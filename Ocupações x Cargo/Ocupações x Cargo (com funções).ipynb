

#### Fazendo o processo antigo, dessa vez usando funções.

Como a utilidade das funções é a de evitar a reutilização ao mesmo tempo que aumenta a manutenbilidade do código, a priori tudo que é repetido pode ser transformado em uma função.



A sintaxe de criar uma função é somente a seguinte:
`def "nome_da_função(parametro1, parametro2...parametroN):`



Assim, resolvi criar uma função chamada `profissões_por_cargo` que recebe o `nome_do_cargo` como um parâmetro que o usuário vai inserir e, consequentemente, alterar o resultado encontrado. Todos os outros termos e operações continuam inalterados.



Este é um caso simples, pois basta dizer ao código que o parâmetro `nome_do_cargo` será utilizado no trecho **[profissoes['Cargo']==*nome_do_cargo])**.



Ao rodar a célula seguinte com, por exemplo, `profissões_por_cargo('Deputado Distrital')` o valor `'Deputado Distrital` será inserido no lugar de `nome_do_cargo` dentro do código da função. O mesmo pode ser dito para qualquer valor encontrado na célula acima, afinal são nomes válidos de cargos políticos segundo a tabela fornecida.



Ao final deste processo as informações coletadas são armazenadas em uma variável chamada `profissões_filtradas` e *retornado*  para que o valor seja usado na execução da função. Caso não fosse *retornado*, o resultado não poderia ser acessado na próxima célula.



```python
def profissoes_por_cargo(nome_do_cargo):
    profissoes_filtradas = (profissoes.loc[profissoes['Cargo']==nome_do_cargo]).sort_values(by = "Quantitativo", ascending = False)
    return profissoes_filtradas
```

```python
profissoes_por_cargo('Deputado Distrital')
```
